```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

###### Continuing visualizations

```python
# pie chart for "price_range"
pie_price_range = df['price_range_loc'].value_counts().sort_index().plot.pie(autopct='%1.1f%%', colors = sns.color_palette('cool', 3))
plt.tight_layout()
plt.ylabel('')
```

```python
# pie chart for "loyalty_flag"
pie_loyalty = df['loyalty_flag'].value_counts().sort_index().plot.pie(autopct='%1.1f%%', colors = sns.color_palette('coolwarm', 3))
plt.tight_layout()
plt.ylabel('')
```

```python
# pie chart for "spending_flag"
pie_spending = df['spending_flag'].value_counts().sort_index().plot.pie(autopct='%1.1f%%', colors = sns.color_palette('spring', 2))
plt.tight_layout()
plt.ylabel('')
```

```python
# export "pie_price_range" Pie Chart
pie_price_range.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'pie_price_range.png'))
```
